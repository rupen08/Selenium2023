## Implicit and Explicit waits

### Implicit wait: 

It will aplly to all the web elements on that web page.

It will NOT WAIT if the element is found and will move to the next one very next moment.

But if the element is not found, it will wait for pre-defined time BEFORE throwing any error.

driver.implicitly_wait(15)


### Explicit wait:

It will be aplly to only SPECIFIC web elements on that web page.

```python
## Implicit wait

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)




class demo_Implicitwaits():
    def iwait(self):
        driver = webdriver.Chrome(options=chrome_options)
        # define the implicit wait here for 15 seconds
        # this wait will apply to all web elements on that page
        # this will wait ONLY if the element is not found OTHERWISE move to next one once it is found
        driver.implicitly_wait(15)
        # to go to the website
        driver.get("https://login.salesforce.com/?locale=ca") 
        driver.maximize_window()
        driver.find_element(By.ID,"username").send_keys("rupen6804@gmail.com")
        
        # passing the wrong ID name, it should throw an error after waiting 5 sec instead of instant
        driver.find_element(By.ID,"passdfvword").send_keys("123456")  
        # it will wait 15 secs before throwing an error      
        driver.quit()

name1 = demo_Implicitwaits()
name1.iwait()
```
